# Task 1 - Plot a line graph to show the difference between heart rate of smokers and non smokers

```python
#Import libraries
import pandas as pd
import matplotlib .pyplot as plt
#read the csv
dataframe = pd.read_csv("framingham.csv")
df = dataframe.dropna()
df
```

```python
#Filter and make a new dataframe for non smokers
non_smokers = df.loc[df['currentSmoker']==0]
non_smokers
```

```python
#Group by age column and find average heart rate at different age
group_non_smokers = non_smokers.groupby('age')['heartRate'].mean().reset_index()
group_non_smokers
```

```python
#Filter and make a new dataframe for smokers
smokers = df.loc[df['currentSmoker']== 1]
smokers
```

```python
#Group by age column and find average heart rate at different age
group_smokers = smokers.groupby('age')['heartRate'].mean().reset_index()
group_smokers
```

```python
#Plot a line graph to show the heart rate of smokers vs non smokers

non_smokers_value = group_non_smokers['age']
non_smokers_label = group_non_smokers['heartRate']
fig=plt.subplots(figsize=(18,7))
plt.plot(non_smokers_value,non_smokers_label, label="non_smokers",linewidth=3.5)

smokers_value = group_smokers['age']
smokers_label = group_smokers['heartRate']
plt.plot(smokers_value,smokers_label, label="smokers",linewidth=3.5)


plt.xlabel('age')
plt.xticks(rotation='vertical')
plt.title('Smokers vs Non Smokers')
plt.legend()
plt.show()
```

# Task 2 - Which age group have high chances of having coronary heart disease in 10 years

```python
#Read the csv
dataframe = pd.read_csv("framingham.csv")
df = dataframe.dropna()
df
#Filter and make a new dataframe for those who has chances of having coronary heart disease in 10 years
coronary_heart_disease = df.loc[df['TenYearCHD']== 1]
coronary_heart_disease
```

```python
#Group by age column and count the rows of TenYearCHD column
group_age=coronary_heart_disease.groupby('age')['TenYearCHD'].count().reset_index()
group_age
```

```python
#Plot a line graph to show total number of people having a chance of coronary heart disease in 10 years
plt.figure(figsize=(20,8))

plt.scatter(group_age['age'], group_age['TenYearCHD'],
           color='blue',label='Ten year CHD',
           alpha=10,
           s = group_age['TenYearCHD']*2 )

plt.xticks(rotation='vertical')
plt.legend()
plt.xlabel("age",size=14)
plt.ylabel("Ten Year CHD",size=14)
plt.title("people having chances of chd after ten years")
```
